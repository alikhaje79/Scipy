# import ndimage for rotate
# import misk for import a picture
# import matplotlib for show picture

```python
from scipy import misc,ndimage
import matplotlib.pyplot as plt
```

```python
face = misc.face()
plt.imshow(face)
plt.show()
```

```python
rotate_face = ndimage.rotate(face , angle = 45)
plt.imshow(rotate_face)
plt.show()
```
