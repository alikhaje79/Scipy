```python
from scipy import ndimage as nd
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_image = np.zeros((256,256))
```

```python
my_image[30 : -30 , 30:-30] = 4
my_image[90 : -90 , 90:-90] = 5
```

```python
my_image = nd.gaussian_filter(my_image , 10)
```

```python
plt.imshow(my_image)
plt.show()
```

```python
sx = nd.sobel(my_image , axis=0 , mode='constant')
sy = nd.sobel(my_image , axis=1 , mode='constant')
```

```python
sob = np.hypot(sx,sy)
```

```python
plt.imshow(sob)
plt.show()
```
